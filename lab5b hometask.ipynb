```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = {
    'Booking Source': [1, 2, 3, 4, 5],
    'Booked': [1, 2, 3, 4, 5]  
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)

plt.xlabel('Booking Source')
plt.ylabel('Booked')
plt.scatter(df['Booking Source'], df['Booked'], color='red', marker='+', label='Data Points')

X = df[['Booking Source']]  
y = df['Booked']           

reg = LinearRegression()
reg.fit(X, y)
print(f"Coefficient: {reg.coef_}")
print(f"Intercept: {reg.intercept_}")
y_pred = reg.predict(X)
plt.plot(df['Booking Source'], y_pred, color='blue', label='Regression Line')

plt.legend()
plt.show()

new_data = pd.DataFrame({'Booking Source': [6, 7, 8]})
predicted_booked = reg.predict(new_data)
for source, prediction in zip(new_data['Booking Source'], predicted_booked):
    print(f"Predicted 'Booked' value for Booking Source {source} is {prediction:.2f}")
```
